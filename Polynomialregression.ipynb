Importing lib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
dataset = pd.read_csv('Data/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Training the linear reg model on whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree =2)
x_poly = poly_reg.fit_transform(X)
```

```python
x_poly
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, Y_train, Y_test = train_test_split(x_poly,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
lin_reg2.fit(X_train,Y_train)
```

```python
y_pred = lin_reg2.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_test, y_pred))
```

# Visualising the linear reg results

```python
plt.scatter(X, y,color='red')
plt.plot(X, lin_reg.predict(X),color='blue')
plt.title('Truth or Buff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# Visualising the Polynomial reg results

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color='blue')
plt.title('Truth or Buff(Polynomial Regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```
