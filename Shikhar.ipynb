```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Data/kc_house_data.csv')
x= dataset.iloc[:,1:-1] #slicing, iloc used for rows
y=dataset.iloc[:,-1].values
```

```python
#removing date col
x=x.drop(columns=['date'])
```

# Importing all Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score, recall_score, f1_score,r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,
    test_size=0.25,
    random_state=42
)
```

```python
print(x_train.shape) #It displays the shape (dimensions) of the training feature dataset. Op is in form of rows and col  Rows: no of training sample, col- no of ip feature
```

```python
print(x_test.shape) #print how much rows train
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

## Linear Regression

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred)# Accurate result
```

```python
r2
```

## Logistic Regression

```python
model=LogisticRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
Lr_pred= model.predict(x_test)
```

```python
r2=r2_score(y_test,Lr_pred)
```

```python
r2
```

### Polynomial Regression

```python
poly= PolynomialFeatures(degree=2)
```

```python
x_train_poly= poly.fit_transform(x_train)
x_test_poly= poly.transform(x_test)
```

```python
poly_model= LinearRegression()
```

```python
poly_model
```

```python
poly_model.fit(x_train_poly,y_train)
```

```python
poly_pred= poly_model.predict(x_test_poly)
```

```python
poly_r2= r2_score(y_test,poly_pred)
```

```python
poly_r2
```

### SVM

```python
svr_model= SVR(kernel='rbf')
```

```python
svr_model
```

```python
svr_model.fit(x_train,y_train)
```

```python
svr_pred= svr_model.predict(x_test)
```

```python
svr_r2= r2_score(y_test,svr_pred)
```

```python
svr_r2
```
