```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
tips=pd.read_csv('Data/tipsdataset.csv')
```

```python
tips
```

```python
#univariate analysis
plt.plot(tips['total_bill'])
plt.show()
```

```python
plt.scatter(tips['size'],tips['total_bill'], s=tips['total_bill'])
plt.title('Scatter')
plt.show()
```

```python
# cat using value counts
#multivariate analysis
plt.pie(tips['gender'].value_counts(),autopct='%1.1f%%') # pie chart used to show the distribution of a single categorical variable
plt.show()
```

```python
#univariate analysis numerical data, bin mean divide data into 4 parts 1-100, 1 part -1-24, 2part - 25-49, 3 part - 50-74, 4 part - 75-100  
plt.hist(tips['tip'], bins=4)
plt.show()
```

```python
sns.relplot(data=tips,x='tip' ,y='size', col ='gender', hue='smoker', style='smoker', size='total_bill')
plt.show()
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
plt.show()
```

```python
sns.histplot(data=tips, x="gender",  kde=True)
sns.histplot(data=tips, x="tip",  kde=True)


```

```python
#distribution plot
sns.displot(data=tips, x="gender", col="time", kde=True)
plt.show()
```
