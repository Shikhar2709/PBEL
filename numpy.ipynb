```python
arr= np.array([1, 2, 3, 4, 5])
print(arr)  # Output: [1 2 3 4 5]
print(type(arr))  # Output: <class 'numpy.ndarray'>
```

```python
print(len (arr) )  # Output: 10

```

```python
## attribute
print(arr.shape)  # Output: (5,)
print(arr.ndim)  # Output: 1
print(arr.size)  # Output: 5
```
